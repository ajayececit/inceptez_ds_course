```python
def missing_percent(train):
    nan_percent= 100*(train.isnull().sum()/len(train))
    nan_percent= nan_percent[nan_percent>0].sort_values()
    return nan_percent
```

```python
## Adding a column to identify the train and test after concatenate 

train['source'] = 'train'
test['source'] = 'test'
```

```python
df = pd.concat([train,test])
print(df.head())
```

```python
## Finding numerical and categorical columns
numerical_columns = df.select_dtypes(include=np.number).drop(columns = ['YearBuilt','YearRemodAdd', 'YrSold', 'SalePrice'])
categorical_columns = df.select_dtypes(exclude = np.number).columns
```

```python
train['LotFrontage']= train['LotFrontage'].fillna(0)
```

```python
# drop the columns which are hacing more than 70% missing 
df.drop(['PoolQC','MiscFeature','Alley','Fence'],axis=1,inplace=True)
```

```python
categorical_columns = df.select_dtypes(exclude = np.number).columns
```

```python
categorical_columns
```

# Fill na with mean and mode

```python
for i in categorical_columns:
  df[i] = df[i].fillna(df[i].mode()[0])

for i in numerical_columns:
  df[i] = df[i].fillna(df[i].mean())
```
