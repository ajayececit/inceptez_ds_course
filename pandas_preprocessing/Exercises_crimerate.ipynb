### Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv('https://raw.githubusercontent.com/Laxminarayen/Inceptz-Batch13-Analytics_and_Python/master/08%20-%20Day%20-%208%20-%20Python%20Quiz%20Session/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
```

### Step 4. What is the type of the columns?

```python
print(crime.dtypes)
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'].astype(str))

crime['Year'] = pd.to_datetime(crime['Year'], format="%Y")
print(crime.dtypes)
print(crime.head())
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime = crime.set_index("Year")
print(crime)
```

### Step 7. Delete the Total column

```python
crime.drop('Total', axis=1, inplace=True)
print(crime.head())
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crime_group = crime.groupby((crime.index.year//10)*10).sum()
```

### Step 9. What is the mos dangerous decade to live in the US?

```python
header_list = list(crime_group.columns)
header_list = header_list[1:]

crime_group['total_crime'] = crime_group.loc[:,header_list].sum(axis=1)
crime_group['crime_rate'] = (crime_group["total_crime"] / crime_group["Population"]) * 100
max_crime = crime_group["crime_rate"]
dangerous_year = max_crime.idxmax()
print(f"dangerous year with max crime rate : {dangerous_year}")
```
